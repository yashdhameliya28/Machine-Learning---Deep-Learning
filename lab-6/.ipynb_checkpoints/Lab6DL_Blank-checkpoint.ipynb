# Decision Tree Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
```

```python
data
```

# Importing the libraries

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
```

# Divide the data into input and output

```python
X = data['data']
```

```python
y = data['target']
```

```python
print(X.shape, y.shape)
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

# Fitting DecisionTreeClassifier   on dataset

```python
clf = tree.DecisionTreeClassifier(random_state=42, criterion='entropy', max_depth=100)
```

```python
clf.fit(X_train, y_train)
```

# Display Decision Tree

```python
from sklearn import tree
```

```python
%matplotlib qt
```

```python
tree.plot_tree(clf)
```

# Predict the x_test 

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
# accuracy of model training
clf.score(X_train, y_train)
```

```python
# accuracy of model testing
clf.score(X_test, y_test)
```

```python
# save model 
import pickle

with open('model.pkl', 'wb') as file:
    pickle.dump(clf, file)
```

```python
# load model from disk

with open('model.pkl', 'rb') as file:
    load_model = pickle.load(file)
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
df = df.replace(0, np.nan)
```

```python
df
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
medi_pre, medi_gluc, medi_bp, medi_skin, medi_ins, medi_bmi = df['Pregnancies'].median(), df['Glucose'].median(), df['BloodPressure'].median(), df['SkinThickness'].median(), df['Insulin'].median(), df['BMI'].median()
```

```python
df['Pregnancies'].fillna(medi_pre, inplace=True)
df['Glucose'].fillna(medi_gluc, inplace=True)
df['BloodPressure'].fillna(medi_bp, inplace=True)
df['SkinThickness'].fillna(medi_skin, inplace=True)
df['Insulin'].fillna(medi_ins, inplace=True)
df['BMI'].fillna(medi_bmi, inplace=True)
```

```python
df.isnull().sum()
```

# Visualize Distributions

```python
sns.pairplot(df, hue='Outcome', kind='reg')
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr)
```

```python
df
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Diabetic':0, 'Non Diabetic':1})
```

```python
df
```

# Divide the data into input and output

```python
X = df.drop('Outcome', axis=1)
```

```python
y = df['Outcome']
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
```

# Create Model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42, max_depth=100)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model.fit(X_train, y_train)
```

# Display Decision Tree

```python
%matplotlib qt
tree.plot_tree(model)
```

# Predict the x_test 

```python
y_pred = model.predict(X_test)
```

# Display Training Accuracy

```python
model.score(X_train, y_train)
```

# Display Test Accuracy

```python
model.score(X_test, y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score
print(classification_report(y_test, y_pred))
```

# Save Model State

```python
import pickle
```

```python
with open('diabetes_decision_model.pkl', 'wb') as file:
    pickle.dump(model, file)
print("Model saved")
```

# Load Model from Disk

```python
with open('diabetes_decision_model.pkl', 'rb') as file:
    load_model = pickle.load(file)
print("Model loaded")
```
