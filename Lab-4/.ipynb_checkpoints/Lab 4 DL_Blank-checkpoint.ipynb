## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2. Import the dataset

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# check datatype
df.info()
```

```python
# check null value 
df.isnull().sum()
```

```python
# check duplicate data
df.duplicated().sum()
```

## Step 3 . Check the State Column

```python
df['State'].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
X=df.drop('Profit', axis=1)
y=df['Profit']
```

```python
# X=df.iloc[::, :4:]
# y=df.iloc[::, 4::]
```

```python
X.head()
```

```python
y.head()
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# encoder=OneHotEncoder(handle_unknown='ignore')
# stat_encoded=encoder.fit_transform(X[['State']])
# stat_encoded
```

```python
X=pd.get_dummies(X, drop_first=True)
X
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(X_train, y_train)
```

## Step 10  Predict the data

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
plt.scatter(X_test['Administration'], y_test, label='Actual')
plt.scatter(X_test['Administration'], y_pred, label='Pridected')
```

```python
plt.scatter(y_pred, y_test)
plt.scatter(y_test, y_pred)
plt.show()
```

## RSS

```python
RSS=np.sum((y_test-y_pred)**2)
```

```python
print(RSS)
```

```python
mse=mean_squared_error(y_test, y_pred)
print(mse)
```

## R Square

```python
from sklearn.metrics import r2_score

r2=r2_score(y_test, y_pred)
print(r2)
```
